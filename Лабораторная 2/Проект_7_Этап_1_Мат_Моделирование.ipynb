## 1. Теоретическое описание задачи

---

### 1.1. Небулярная гипотеза

#### Изучение механизма формирования планет из газопылевого облака

> Согласно теории, газопылевое облако, из которого позднее образовались планеты и звезда, имело **ненулевой момент импульса**, то есть вращалось.
Суммарный момент импульса в замкнутой системе сохраняется.

По мере гравитационного сжатия облака происходило следующее:
-

1.  Расстояние всех его частей от оси вращения сокращалось
2.  Скорость вращения сгущающегося облака увеличивалась
3.  В плоскости, перпендикулярной оси вращения, сжатие происходило медленнее
4.  Облако, бывшее изначально шаровидным, становилось все более плоским


> **Ключевой принцип:** Сохранение момента импульса при гравитационном сжатии приводит к формированию вращающегося диска.

---

#### Анализ этапов эволюции: коллапс → диск → аккреция → звезда


| № | Этап | Описание | Ключевые процессы |
|:--:|------|----------|-------------------|
| 1 | **Коллапс** | Гравитационное сжатие газопылевого облака под действием собственной гравитации | • Гравитационная неустойчивость<br>• Отделение колец вещества на периферии |
| 2 | **Формирование диска** | Облако сплющивается в протопланетный диск из-за сохранения момента импульса | • Сплющивание в плоскости вращения<br>• Формирование дискообразной структуры |
| 3 | **Аккреция** | Отделившиеся кольца вещества сгущаются в планеты | • Слипание газопылевых облаков<br>• Подчинение третьему закону Кеплера* |
| 4 | **Формирование звезды** | Центральное облако продолжает сжиматься, формируя протозвезду | • Сжатие центральной области<br>• Разогрев и зажигание термоядерных реакций |

> Скорость обращения частиц убывает обратно пропорционально корню квадратному из расстояния до центра: $v \sim 1/\sqrt{r}$

---

### 1.2. Физические процессы

Для численной модели рассматриваются следующие процессы:

| Процесс | Описание | Математическое описание |
|---------|----------|-------------------------|
| **Гравитационная неустойчивость** | Начальные возмущения плотности в диске, приводящие к формированию сгущений | Критерий Джинса: $M_J > \frac{5kT}{G\mu m_H} R$ |
| **Аккреция** | Рост массы протопланет за счет слипания частиц при столкновениях | $\frac{dm}{dt} = \pi R^2 \rho v_{rel}$<br>При слиянии: $m_{new} = m_i + m_j$, $R_{new} = \sqrt[3]{R_i^3 + R_j^3}$ |
| **Миграция** | Изменение орбит планет под влиянием взаимодействия с газом диска | Типы миграции: I тип (для малых планет) и II тип (для крупных планет) |

---

### 1.3. Аналогия с поведением людей

Интересно отметить, что уравнения движения частиц в протопланетном диске математически аналогичны уравнениям, описывающим движение людей в стрессовых ситуациях (например, при эвакуации из помещения).

**Сравнительный анализ:**

| Компонент | В модели образования планет | В модели поведения людей |
|-----------|----------------------------|--------------------------|
| **Направленная сила** | Гравитационное притяжение к звезде | Стремление выйти в дверь $F^s_i = m(v^* - v_i)/\tau$ |
| **Сила отталкивания** | Предотвращает слияние при высоких скоростях | Психологическое и физическое отталкивание $F^r_{ij} = F_0 \left( \frac{r_0}{r_{ij}} - 1 \right)^n \frac{\mathbf{r}_{ij}}{r_{ij}}$ |
| **Сила трения** | Диссипация энергии при столкновениях | Трение при контакте $F^f_{ij} = \mu F^r_{ij}$ |
| **Итоговое уравнение** | $\frac{d\mathbf{v}_i}{dt} = \mathbf{F}^g_i + \sum \mathbf{F}^r_{ij} + \sum \mathbf{F}^f_{ij}$ | $\frac{d\mathbf{v}_i}{dt} = \mathbf{F}^s_i + \sum \mathbf{F}^r_{ij} + \sum \mathbf{F}^f_{ij}$ |

**Ключевое различие:**
- Газопылевые сгустки подчиняются только физическим законам
- Люди обладают свойством, которого нет у газопылевых сгустков — они **живые**, поэтому, если на человека действует слишком большая сила, это может привести к травме, потере сознания и даже смерти

> Математическое описание этих, казалось бы, разных систем приводит к одинаковым уравнениям движения, что демонстрирует универсальность физических законов и методов математического моделирования.


## 2. Описание модели

---

### 2.1. Допущения модели

Для упрощения расчетов и создания вычислительной модели вводятся следующие допущения:

| № | Допущение | Обоснование |
|:--:|-----------|-------------|
| 1 | **Аксиально-симметричный диск** | Протопланетный диск обладает осевой симметрией относительно оси вращения, что позволяет упростить расчеты |
| 2 | **Звезда массой 1 M☉ в центре** | Моделируется система, аналогичная Солнечной, где центральное тело содержит основную массу системы |
| 3 | **2D-модель (плоскость диска)** | Вещество диска сконцентрировано в тонком слое, что позволяет пренебречь вертикальной структурой |
| 4 | **Абсолютно неупругие столкновения** | При столкновении частицы слипаются, теряя относительную скорость (моделирует коагуляцию пыли) |

> **Важно:** Данные допущения позволяют снизить вычислительную сложность задачи, сохраняя при этом основные физические закономерности процесса формирования планет.

---

### 2.2. Алгоритм симуляции

Модель реализуется пошагово с использованием метода частиц:

| Шаг | Этап | Действия | Описание |
|:--:|------|----------|----------|
| 1 | **Инициализация** | Задание начальных параметров | • Распределение частиц в диске<br>• Начальные массы и радиусы<br>• Начальные скорости (кеплеровское вращение)<br>• Параметры звезды |
| 2 | **Расчет сил** | Вычисление гравитационного взаимодействия | • Силы между частицами: $F = G\frac{m_i m_j}{r_{ij}^2}$<br>• Сила притяжения к звезде: $F = G\frac{M m_i}{r_i^2}$ |
| 3 | **Слияние** | Объединение частиц при столкновении | • Проверка условия: $r_{ij} < R_i + R_j$<br>• Новая масса: $m_{new} = m_i + m_j$<br>• Новый радиус: $R_{new} = \sqrt[3]{R_i^3 + R_j^3}$<br>• Новая скорость: $\mathbf{v}_{new} = \frac{m_i\mathbf{v}_i + m_j\mathbf{v}_j}{m_i + m_j}$ |
| 4 | **Обновление** | Пересчет координат и скоростей | • Интегрирование уравнений движения<br>• Обновление позиций: $\mathbf{r}(t+\Delta t)$<br>• Обновление скоростей: $\mathbf{v}(t+\Delta t)$ |
| 5 | **Визуализация** | Вывод текущего состояния | • Отображение частиц на плоскости<br>• Графики распределения масс<br>• Траектории движения |
